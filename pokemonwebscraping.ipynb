```python
import scrapy
import json
from scrapy.crawler import CrawlerProcess
```

```python

class jsonSpider(scrapy.Spider):
  name = "pokemon151Spider"
  base_url = 'https://www.pricecharting.com/console/pokemon-scarlet-&-violet-151?sort=&release-date=2024-07-29&cursor={}&format=json'

  def start_requests(self):
      start_url = self.base_url.format(0)
      yield scrapy.Request(url=start_url, callback=self.parse)


  def parse(self,response):
    data = json.loads(response.text)

    for product in data.get("products",[]):
      yield{
          "id" : product.get("id"),
          "image": product.get("imageUri"),
          "name": product.get("productName"),
          "ungraded_price": product.get("price1")

      }

    next_cursor = data.get('cursor')
    if next_cursor and next_cursor != "null":
      next_page_url = self.base_url.format(next_cursor)
      yield scrapy.Request(url=next_page_url, callback=self.parse)

process = CrawlerProcess(settings = {
    "FEEDS": {
        "output.csv" : {
            "format" : "csv",
            "fields" : ["id", "image", "name", "ungraded_price"]
        },
    },
})
process.crawl(jsonSpider)
process.start()
```
