```python
import scrapy
```

```python
class CardSpider(scrapy.Spider):
  name = "pokemonSpider"
  start_urls = ['https://www.pricecharting.com/console/pokemon-scarlet-&-violet-151']

  def parse(self, response):
      title = response.xpath('//*[@id="console-page"]//*[@id="games_table"]/tbody/tr//*[@class="title"]/a/text()').getall()
      ungraded = response.xpath('//*[@id="console-page"]//*[@id="games_table"]/tbody/tr//*[@class="price numeric used_price"]/span/text()').getall()
      for name, price in zip(title, ungraded):
        yield{"name" : name,
              "ungraded price" : price}
```

```python
from scrapy.crawler import CrawlerProcess

process = CrawlerProcess(settings = {
    "FEEDS": {
        "output.csv" : {
            "format" : "csv",
            "fields" : ["name", "ungraded price"]
        },
    },
})
process.crawl(CardSpider)
process.start()
```
